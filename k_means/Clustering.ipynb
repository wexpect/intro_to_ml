```python
def Draw(pred, features, poi, mark_poi=False, name="image.png", f1_name="feature 1", f2_name="feature 2"):
    """ some plotting code designed to help you visualize your clusters """

    ### plot each cluster with a different color--add more colors for
    ### drawing more than five clusters
    colors = ["b", "c", "k", "m", "g"]
    for i, v in enumerate(pred):
        plt.scatter(features[i][0], features[i][1], color = colors[v], marker='x')

        ### if you like, place red stars over points that are POIs (just for funsies)
        if mark_poi and v:
            plt.scatter(features[i][0], features[i][1], color="r", marker="+")
            
#     plt.scatter(200320.46218487, 1507501.63865546, color='r', marker='v')
#     plt.scatter( 573218.8,  23904558., color='r', marker='v')   
        
#         plt.scatter(1.85738464e+05, 1.91715291e+06, color='r', marker='v')
#         plt.scatter(1.07232100e+06, 3.43483840e+07, color='r', marker='v')   

            
    plt.xlabel(f1_name)
    plt.ylabel(f2_name)
    plt.savefig(name)
    plt.show()
```

```python
Draw(pred, finance_features, poi, mark_poi=True, name="clusters.pdf", f1_name=feature_1, f2_name=feature_2)
```
